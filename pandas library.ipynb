# Creating a DataFrame from a dictionary of series

```python
import pandas as pd
import numpy as np
series_df = pd.DataFrame({
    'A': 1000.,
    'B': pd.Timestamp('20190426'),
    'C': pd.Series(5, index=list(range(4)), dtype='float32'),
    'D': np.array([3] * 4, dtype='int32'),
    'E': pd.Categorical(["this", "is", "a", "test"]),
    'F': 'I love Pandas',
    'G': 'Here is my code'
})
print(series_df)
```

# Loading a dataset into Pandas DataFrame

In this example, I have created a dataframe by loading CSV data into it with a simple read_csv() function in Pandas.

```python
import pandas as pd
cols = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
    'marital_status', 'occupation', 'relationship', 'ethnicity', 'gender','capital_gain','capital_loss','hours_per_week','country_of_origin','income']
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',names=cols)
print(df)
```

# Selecting rows and columns in the dataframe

.iloc — Selects rows and columns by names

```python
df.iloc[10]              - Selects a row
df.iloc[0:10]            - Selects 10 rows
df.iloc[10:15]           - Selects a range of rows 
df.iloc[-2:]             - Selects the last 2 rows
df.iloc[::2, 3:5].head() - Selects every other row in columns 3-5
```

.loc — Selects rows and columns by index position

```python
df.loc[0:2, 'age']  - Selects the first 3 records from 'age' column
df.loc[df.age > 50, ['age', 'education', 'occupation', 'gender', 'income']]  - Filters data based on the given condition
df.loc[df.gender == 'Male', ] - Filters on Gender
(df.gender == 'Male') & (df.age >= 30) & (df.age <= 40) - Filters on gender and age between 30 and 40
```

# Find, fill and drop nulls

```python
# Finding null values in the 'age' column
df[df.age.isnull()]
```

How to fill null values using mean, median of the column or with a value etc.

```python
# Fill nulls
null_df = pd.DataFrame([1, 2, 4, np.nan], columns=['sample_column'])
print(null_df)
```

```python
# you can also fill nulls with a value or string
null_df['sample_column'].fillna(1000)
```

```python
null_df.fillna(null_df.sample_column.median(), inplace=True)
null_df
```

# Groupby

```python
# Groups the data by the 'relationship' column
df.groupby('relationship').count()
```

```python
# How to groupby column and apply a function like sum, count, or mean
df.groupby(['education']).mean()
```

```python
# can use the aggs function to aggregate columns separately
gb = df.groupby(['income', 'education'])
gb_aggs = gb.agg({'age': 'mean', 'capital_gain': 'sum'})
gb_aggs.sample(10)
```

# Sort

1. sort_index() to sort by index
2. sort_values() to sort by values

```python
# groupby income and workclass and then sort by their mean age within each data block
df_grouped = df.groupby(['income', 'workclass']).mean().sort_values(
    'age', ascending=True)
df_grouped
```

# Merge

```python
Sample_df1 = pd.DataFrame({'EmpName': ['Jack', 'Jill', 'Alan'],
                           'EmpNo': ['12', '13', '14'],
                           'EmpAge': ['29', '28', '23']})
 
Sample_df2 = pd.DataFrame({'EmpName': ['Jack', 'Jill', 'Alan'],
                           'EmpSalary': ['1500', '1899', '900']})
```

```python
Sample_df1
Sample_df2
```

```python
Merged_df = pd.merge(Sample_df1, Sample_df2, on='EmpName')
Merged_df
```

# Concatenation

Concatenation in Pandas can be using the function concat().

```python
print(pd.concat([Sample_df1, Sample_df2]))
```
